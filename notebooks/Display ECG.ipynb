ECG signal

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

Import an EGC data

```python
# load the data
ECG_data = np.loadtxt('../data/ECGu.txt')

# look at the data
print(ECG_data)
print(ECG_data.shape)
```

Convert raw data

```python
#create a time vector t at the same lenght as ECG and with a sampling rate of 1000 Hz
fs = 1000  # Sampling frequency in Hz
t = np.arange(0, ECG_data.shape[0]) / fs

```

```python
# rescale the ECG to mV knowing that the conversion is 1024 microV per unit
#ECG = ECG * 1024 / 1000  # Convert to mV

analog_to_digital_conversion_gain = 1024 # microvolt per unit
microvolt_to_mV = 1000
ECG = ECG_data* analog_to_digital_conversion_gain / microvolt_to_mV 
```

Diplay ECG signal

```python
plt.figure(figsize=(7,5))   #Choose a good size to read quickly
plt.grid(True)              #Place a grid
plt.plot(t,ECG)          #Place x and y 
plt.title("ECG Signal")     #Name title, x and y 
plt.xlabel("Time (s)")      
plt.ylabel("ECG (mV)")
```

Create a realistic ECG plot

```python
plt.figure(figsize=(12,5))   
plt.grid(True,which='major', color='red',linestyle='-', linewidth= 0.8) 
plt.minorticks_on()
plt.grid(True,which='minor', color='red',linestyle='--', linewidth= 0.3)
plt.plot(t,ECG[:,0])        
plt.title("ECG Sigal I")     
plt.xlabel("Time (s)")      
plt.ylabel("ECG (mV)")

plt.figure(figsize=(12,5))  
plt.grid(True,which='major', color='red',linestyle='-', linewidth= 0.8)
plt.minorticks_on()
plt.grid(True,which='minor', color='red',linestyle='--', linewidth= 0.3)              
plt.plot(t,ECG[:,1])         
plt.title("ECG Signal II")      
plt.xlabel("Time (s)")
plt.ylabel("ECG (mV)")

plt.figure(figsize=(12,5))  
plt.grid(True,which='major', color='red',linestyle='-', linewidth= 0.8)
plt.minorticks_on()
plt.grid(True,which='minor', color='red',linestyle='--', linewidth= 0.3)              
plt.plot(t,ECG[:,2])          
plt.title("ECG Signal III")     
plt.xlabel("Time (s)")
plt.ylabel("ECG (mV)")
```
